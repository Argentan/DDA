```python
import pandas as pd
```

```python
path=r".\Venta QUILMES 2016 2017 2018 x Cliente (1).xlsx"

data = pd.read_excel(path)
data
```

```python
data.dtypes
```

```python
data = data.drop("Marca", axis=1)
```

```python
import numpy as np
```

```python
data = pd.crosstab(data["Código Cliente"], data["Año/Mes"], values=data["Suma de Cantidad Total en HL"], aggfunc=np.sum)
data
```

```python
data.isnull().sum()
```

```python
train = data[data.columns[:-2]]
train.columns
```

```python
test = data[data.columns[1:-1]]
test.columns
```

```python
target_col_train = train.columns[-1]
target_col_test = test.columns[-1]

deltas = train.T.pct_change().T.add_suffix('_delta')
train = train.join(deltas.drop(deltas.columns[[0, -1]], axis=1))

deltas = test.T.pct_change().T.add_suffix('_delta')
test = test.join(deltas.drop(deltas.columns[[0, -1]], axis=1))
```

```python
from lightgbm import LGBMRegressor, LGBMClassifier
```

```python
res = []
for i in range(5):
    X_train = train.sample(frac=0.8).drop(target_col_train, axis=1)
    y_train = train.loc[X_train.index][target_col_train].fillna(0)
    X_valid = train.drop(X_train.index).drop(target_col_train, axis=1)
    y_valid = train.loc[X_valid.index][target_col_train].fillna(0)
    learner = LGBMRegressor(n_estimators=10000)
    learner.fit(X_train, y_train, early_stopping_rounds=10,
                eval_set=[(X_train, y_train),
                         (X_valid, y_valid)])
    volumen = learner.predict(test.drop(target_col_test, axis=1))
    learner = LGBMClassifier(n_estimators=10000)
    learner.fit(X_train, (y_train > 0).astype(int), early_stopping_rounds=10,
                eval_set=[(X_train, (y_train > 0).astype(int)),
                         (X_valid, (y_valid> 0).astype(int))])
    probabilidad = learner.predict_proba(test.drop(target_col_test, axis=1))[:, -1]
    res.append(volumen * probabilidad)
```

```python
import numpy as np
res = np.concatenate([r[:, None] for r in res], axis=1).mean(axis=1)
resultados = test.copy()
resultados["preds"] = res
resultados.to_excel("resultados.xlsx")
```
