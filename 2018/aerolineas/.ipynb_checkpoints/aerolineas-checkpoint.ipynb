```python
import pandas as pd
```

```python
path = "./PRUEBA.html"
data = pd.read_html(path)[0]
data
```

```python
data.dtypes
```

```python
columnas_malas = ["col1", "col2"]
data.drop(columnas_malas, axis=1, inplace=True)
```

```python
for c in ["cat_col1", "cat_col2"]:
    data[c] = data[c].astype("category")
```

```python
for c in data.select_dtypes("object"):
    data[c] = data[c].astype("category")
data.dtypes
```

```python
from lightgbm import LGBMClassifier, LGBMRegressor
```

```python
data.sample?
```

```python
train = data.sample(frac=0.9)
test = data.drop(train.index)
```

```python
target_col = "outcome"
res = []
for i in range(5):
    X_train = train.sample(frac=0.8).drop(target_col, axis=1)
    y_train = train.loc[X_train.index][target_col]
    X_valid = train.drop(X_train.index).drop(target_col, axis=1)
    y_valid = train.loc[X_valid.index][target_col]
    learner = LGBMClassifier(n_estimators=10000)
    learner.fit(X_train, y_train, early_stopping_rounds=10,
                eval_set=[(X_train, y_train),
                         (X_valid, y_valid)])
    res.append(learner.predict(test.drop(target_col, axis=1)))
```

```python
res = pd.concat(res).mean(axis=1)
test = test.copy()
test["preds"] = res
test.to_excel("resultados.xlsx")
```
